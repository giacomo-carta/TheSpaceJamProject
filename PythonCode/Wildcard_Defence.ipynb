```python
# We are going to draft our first wildcard: a DEFENSIVE specialist!
# First of all we are going to import the LeagueHustleStatsPlayer endpoint

from nba_api.stats.endpoints import LeagueHustleStatsPlayer

params = {
    "season": "2023-24",
    "season_type_all_star": "Regular Season",
    "per_mode_time": "PerGame"
}

hustle_stats_player = LeagueHustleStatsPlayer(**params)
df_defWC = hustle_stats_player.get_data_frames()[0]

df_defWC.head()
```

```python
# Let's have a look at the attributes that we are going to be working with

df_defWC.columns
```

```python
# Let's only select players who have at least 20 games played and more than 30 minutes per game

df_defWC2 =  df_defWC[(df_defWC.G >= 20) & (df_defWC.MIN >= 30)].copy()
```

```python
# We are going to select some of the most important defensive stats and add them together to find the best defensive players in the league

df_defWC2['def_par'] = (df_defWC2['CONTESTED_SHOTS']*0.5) + df_defWC2['DEFLECTIONS'] + df_defWC2['LOOSE_BALLS_RECOVERED'] + df_defWC2['BOX_OUTS']

df_defPar = df_defWC2[['PLAYER_NAME', 'def_par', 'CONTESTED_SHOTS', 'DEFLECTIONS', 'LOOSE_BALLS_RECOVERED', 'BOX_OUTS']]

df_defWC2[['PLAYER_NAME', 'def_par', 'CONTESTED_SHOTS', 'DEFLECTIONS', 'LOOSE_BALLS_RECOVERED', 'BOX_OUTS']].sort_values(by='def_par',ascending=False).head(20).round(1)
```

```python
df_defPar.to_csv('DefWildcardCleaned.csv', index=False)
```
